### 문자열에서 자주 사용하는 함수

```python
# 동전 교환 프로그램 만들기 (최소한의 동전 개수로)
money = int(input('동전으로 교환할 금액 입력하세요:'))
print('-'*50)
print(f"500원짜리 ==> {money // 500}개")
money %= 500
print(f"100원짜리 ==> {money // 100}개")
money %= 100
print(f"50원짜리 ==> {money // 50}개")
money %= 50
print(f"10원짜리 ==> {money // 500}개")
money%= 10
print(f"잔돈 ==>  {money}개")
```

```python
print(f"500원짜리 ==> {}개")
#동전 교환 프로그램 만들기 (최소한의 동전의 개수로)
money,c500,c100,c50,c10=0,0,0,0,0 #전체변수를 포기화
money = int(input("교환할 돈은 얼마? : ")) #money값을 설정해주는 코드
c500= money //500 #입력한 money값을 500으로 나누어 몫을 저장
money = money%500 #500으로 나눈 나머지값을 저장
c100 = money // 100 #입력한 money값을 100으로 나누어 몫을 저장
money%=100 #100으로 나눈 나머지값을 저장
c50 = money // 50 #입력한 money값을 50 으로 나누어 몫을 저장
money%=50 #50으로 나눈 나머지값을 저장
c10= money // 10 #입력한 money값을 10으로 나누어 몫을 저장
money%=10 #10으로 나눈 나머지값을 저장
print("500원짜리 ==> %d"%c500) #밑에 코드들은 그 결과값을 뿌려주는 코드입니다
print("100원짜리 ==> %d"%c100)
print("50원짜리 ==> %d"%c50)
print("10원짜리 ==> %d"%c10)
print("나머지돈 ==> %d"%money) #나누지 못한 값을 표현해줍니다


```
