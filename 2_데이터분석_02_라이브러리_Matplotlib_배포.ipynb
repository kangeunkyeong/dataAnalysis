#### 1.1 pyplot
- 맷플롯립을 이용할 때 가장 기본이 되는 객체
- 파이플롯(pyplot) 위에 그림(figure) 객체를 올리고 그 위에 그래프에 해당하는 축(axes)을 올림
- 그림 위에 축을 여러 장 올리면 여러 개의 그래프 작성

```python
# matplotlib 모듈 호출
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumBarunGothic'
plt.rcParams['figure.figsize'] = (5, 3)
x1 = range(100)
x2 = range(0,200,2)
y = range(100)
plt.plot(x1, y,
         color='red', linestyle='dotted', linewidth='5', label='x1')
plt.plot(x2, y, label='x2')
plt.title('선그래프')
plt.grid()
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.legend()
plt.show()
```

```python
# numpy 모듈 호출

import numpy as np

x_1 = range(100)
y_1 = [np.cos(x) for x in x_1]
y_2 = [np.sin(x) for x in x_1]

plt.plot(x1, y_1)
plt.plot(x1, y_2)
plt.show()
```

#### 1.2 그림과 축
- fig : 그림은 그래프를 작성하는 밑바탕이 됨
- ax  : 축은 실제로 그래프를 작성하는 공간

```python
x_1 = range(100)
y_1 = [np.cos(x) for x in x_1]

# (1) figure와 axes 객체 할당
fig, ax = plt.subplots()


# (2) plot 함수를 사용하여 그래프 생성
ax.plot(x_1, y_1)
ax.set(title='코사인 그래프',
       xlabel='x_1',
       ylabel='y_1')

# (3)그래프 제목,X축 라벨,Y축 라벨 설정


# (4) 그래프 출력
plt.show
```

### 하나의 그래프를 쪼개서 여러 개로 구성할 경우

```python
# (1) figure 반환
fig = plt.figure()

# (2) figure의 크기 지정
fig.set_size_inches(4, 2)


# (3) 첫 번째 그래프 생성
ax_1 = fig.add_subplot(1,2,1) # 1행 2열 첫번째


# (4) 두 번째 그래프 생성
ax_2 = fig.add_subplot(1,2,2)


# (5) 첫 번째 그래프 설정
ax_1.plot(x_1, y_1, c='b')


# (6) 두 번째 그래프 설정
ax_2.plot(x_1. y_2, c='g')

plt.show()
```

#### 1.3 서브플롯 행렬

### 3.3 누적 막대그래프(stacked bar graph)
- 누적 막대그래프(stacked bar graph) : 데이터를 밑에서부터 쌓아올려 데이터를 표현

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])   ### shape
print(X)

for i in range(3):
    plt.bar(X, data[i],
            bottom = np.sum(data[:i], axis=0),
            color = color_list[i],
            label=data_label[i])

plt.legend()
plt.show()









```

```python
import pandas as pd

df = pd.DataFrame(data) ## 2차원의 행렬데이터여야 데이타프레임이 나옴

df.T.plot(kind='bar', stacked=True, figsize=[5, 3])
plt.legend()
plt.show(['A', 'B', 'C'])
df
```

#### 3.2 다양한 범주형 데이터를 나타내는 패싯그리드
- 패싯그리드(FacetGrid) : 그래프의 틀만 제공하여 적당한 그래프를 그려주는 클래스

## <a name="05.3플롯리">05.3 플롯리</a>  ## 요즘 많이 사용하는

### 1. 플롯리의 특징
- 플롯리(plotly) : 비즈니스 인텔리전스(Business Intelligence) 대시보드로 개발된 도구     ### BI   
- 맷플롯립이나 시본은 데이터 분석가들이 데이터의 형태나 분포를 살피기 위해 코드로 사용하는 도구 ### 인사이트를 주기 위한
- 문법은 맷플롯립이나 시본과 유사
- 인터랙션 그래프를 지원
 - - 생성된 그래프에 마우스 커서를 올리면 데이터를 볼 수 있음 (인터랙션 그래프)

```python
import plotly.express as px

df = px.data.iris()     # iris는 판다스 데이터프레임
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
df.head()
```

### 좌표 그래프(coordinates plot)
- 데이터 간 관계를 표현(시본은 제공하지 않지만 플롯리에서 제공하는 기능)

```python
fig = px.parallel_coordinates(df, color="species_id",
        labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                color_continuous_scale=  px.colors.diverging.Tealrose, color_continuous_midpoint=2)
fig.show()
```

## <a name="예제:전세계행복지수">예제: 전세계행복지수 분석</a>

### 1. 파일의 데이터 읽어오기
- Ladder score : 행복지수

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

filename = './world-happiness-report-2021.csv'
happiness_data = pd.read_csv(filename)
happiness_data.head()
```

### 2.데이터 살펴보기

```python
happiness_data.info()
```

```python
happiness = happiness_data.iloc[:,[0,1, 2,6,7,8,9,10,11]]
happiness.head(5)
```

### 데이터 통계치 살펴보기
* 전체 나라 수, 행복지수의 최댓값, 행복지수 평균, 행복지수 최솟값

```python
happiness.describe()   # 기술통계 정보(기술통계만, 숫자데이터만 보는 것, 컨트리네임 등 안 나옴)
```

- **행복지수(Ladder score) 속성을 기준으로 내림차순 정렬: 특정 국가 조회**

```python
#'행복지수' 속성을 기준으로 내림차순 정렬 후, 국가명(Country name)이 'South Korea'인 행을 찾아 출력
happiness = happiness.sort_values('Ladder score', ascending=False)  ## 래더스코어, 내림차순으로 솔트밸류=정렬해줘, 그걸 다시 해피니스로 저장할게 ##정렬이유는 순위보려고
happiness.loc[happiness['Country name']=='South Korea']  ### 그 데이터 중에 한국인 거 가져와줘
```

- **가로막대형그래프 시각화** : 행복지수가 높거나, 낮은 국가

```python
happinessFilter = happiness[(happiness.loc[:,'Ladder score']>=7.5) |\   ### loc는 컬럼이름 가져오는 것, 더블유표시는 또는의 의미
                            (happiness.loc[:,'Ladder score']<=3.5)]
sns.barplot(x='Ladder score', y='Country name', data=happinessFilter, palette='coolwarm')
```

### treemap 시각화
* treemap
  * 계층(트리구조)을 이루는 데이터 전체와 일부분 간의 관계를 파악하기 위한 시각화 기법
  * 범주간의 정확한 비교보다 큰 특징을 살펴볼 때 주로 사용

* `px.treemap(data_frame=데이터프레임객체, path=[부모열,자식열], values=열속성, color=열속성)`
  * path는 [부모,자식] 순서로 작성
  * values는 사각형 크기가 나타내는 속성
  * color는 색상으로 표현하는 속성

```python
fig = px.treemap(data_frame=happiness,
                 path=['Regional indicator','Country name'],
                 values='Ladder score',
                 color='Ladder score')
fig.show()
```

```python
# 그래프 제목 넣기
fig = px.treemap(data_frame=happiness,
                 path=[px.Constant('world'),'Regional indicator','Country name'],
                 values='Ladder score',
                 color='Ladder score')
fig.update_layout(title='나라별 행복 지수', title_x=0.5, width=900, height=900)
fig.show()
```

### sunburst 시각화
* px.sunburst(data_frame=데이터프레임객체, path=[부모열, 자식열], values=열속성, color=열속성)

```python
fig = px.sunburst(data_frame=happiness,
                  path=['Regional indicator', 'Country name'],
                  values='Ladder score',
                  color='Ladder score')
fig.update_layout(title='나라별 행복 지수', title_x=0.5, width=1200, height=900)
fig.show()
```
